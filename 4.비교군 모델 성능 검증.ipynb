```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open('result/target_alleee_final.pkl', 'rb') as f:
    target_alleee = pickle.load(f)
    
with open('result/target_alleee_final_yes_reward.pkl', 'rb') as f:
    target_alleee_1 = pickle.load(f)
    
with open('result/target_alleee_final_no_reward.pkl', 'rb') as f:
    target_alleee_2 = pickle.load(f)
    
with open('result/target_alleee_final_no_reward_spare.pkl', 'rb') as f:
    target_alleee_3 = pickle.load(f)
```

```python
def moving_average(lst):
    window = 5
    ret = []
    for i in range(len(lst)):
        if i < window:
            ret.append(sum(lst[:i+1])/(i+1))
        else:
            ret.append(sum(lst[i-window+1:i+1])/window)
    return ret
```

# ENV ALL 환경

```python
plt.figure(figsize=(30,10))



pt = np.array([0]*100)
ps = np.array([0]*100)
po = np.array([0]*100)

for i in range(len(target_alleee)):
    pt += np.array(target_alleee[i][9]) # dense reward model
    ps += np.array(target_alleee[i][10]) # dense no reward model
    po += np.array(target_alleee[i][11]) # spare reward
    
    

pt = (pt/len(target_alleee)) 
pt = moving_average(pt)
ps = (ps/len(target_alleee)) 
ps = moving_average(ps)
po = (po/len(target_alleee))
po = moving_average(po)


plt.plot(range(len(pt)), pt, linewidth=1, alpha=1,marker = '^',markersize= 10, color='r', label='Dense Reward Function')


plt.plot(range(len(pt)), ps, linewidth=1, alpha=1,markersize= 10,color='g',label='Dense without Reward Function')


plt.plot(range(len(pt)), po, linewidth=1, alpha=1, markersize= 10,color='b',  label='Sparse Reward Function')

plt.legend(fontsize = 25)

plt.title("Comparison of Different Reward Functions")
plt.xlabel("Step")
plt.ylabel("Value")

plt.yticks(range(1, 11))
plt.gca().invert_yaxis()
```

```python
import numpy as np
import matplotlib.pyplot as plt


plt.figure(figsize=(30,10))

pt = np.array([0]*100)
ps = np.array([0]*100)
po = np.array([0]*100)

pt_list = []
ps_list = []
po_list = []

for i in range(len(target_alleee)):
    pt += np.array(target_alleee[i][9]) # dense reward model
    ps += np.array(target_alleee[i][10]) # dense no reward model
    po += np.array(target_alleee[i][11]) # spare reward
    
    pt_list.append(target_alleee[i][9])
    ps_list.append(target_alleee[i][10])
    po_list.append(target_alleee[i][11])

pt_std = np.std(pt_list, axis=0)
ps_std = np.std(ps_list, axis=0)
po_std = np.std(po_list, axis=0)

pt = (pt/len(target_alleee))
pt = moving_average(pt)
ps = (ps/len(target_alleee))
ps = moving_average(ps)
po = (po/len(target_alleee))
po = moving_average(po)

plt.plot(pt, color='r', marker='^', markersize=10, label='Dense Reward Function')
plt.fill_between(range(len(pt)), pt - pt_std, pt + pt_std, color='r', alpha=0.3)

plt.plot(ps, color='g', markersize=10, label='Dense without Reward Function')
plt.fill_between(range(len(ps)), ps - ps_std, ps + ps_std, color='g', alpha=0.3)

plt.plot(po, color='b', markersize=10, label='Sparse Reward Function')
plt.fill_between(range(len(po)), po - po_std, po + po_std, color='b', alpha=0.3)

plt.legend(fontsize = 25)
plt.title("Comparison of Different Reward Functions with Variance")
plt.xlabel("Step")
plt.ylabel("Value")
plt.yticks(range(0, 12))
plt.gca().invert_yaxis()
plt.show()

```

```python
plt.figure(figsize=(30,10))


plt.plot(moving_average(pt_std), color='r',  markersize=10, label='Dense Reward Function')
plt.plot(moving_average(ps_std), color='g', markersize=10, label='Dense without Reward Function')
plt.plot(moving_average(po_std), color='b', markersize=10, label='Sparse Reward Function')
plt.legend(fontsize = 25)
```
